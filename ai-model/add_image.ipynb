```python
import pandas as pd
from bs4 import BeautifulSoup
from selenium import webdriver
from selenium.webdriver import ChromeOptions
import time

# Initialize WebDriver
options = ChromeOptions()
options.add_argument("--headless=new")  # Headless mode for faster execution
driver = webdriver.Chrome(options=options)

def format_query(recipe_name):
    """Convert recipe name to search query format (replace spaces with '+')."""
    return recipe_name.replace(" ", "+")

def parse_image_urls(recipe_name, site):
    """Search for a recipe, scrape the first image URL from the specified site, and return it."""
    
    query = format_query(recipe_name)  # Convert name to query format
    
    if site == "recipetineats":
        search_url = f"https://www.recipetineats.com/?s={query}"
        classes, location, source = "entry-image-link", "img", "src"
    elif site == "allrecipes":
        search_url = f"https://www.allrecipes.com/search?q={query}"
        classes, location, source = "img-placeholder", "img", "src"
    else:
        return None  # If the site is unknown
    
    driver.get(search_url)
    time.sleep(3)  # Wait for page to load
    
    content = driver.page_source
    soup = BeautifulSoup(content, "html.parser")
    
    try:
        a = soup.find_all(attrs={"class": classes})[0]
        name = a.find(location)
        return name.get(source) if name else None
    except (IndexError, AttributeError):
        return None  # Return None if no image is found

# Load DataFrame
df = pd.read_csv("recipes_sample.csv")

# Process only the first 10 rows
def get_image_link(recipe):
    """Try to get image from RecipeTinEats first, if fails, try AllRecipes."""
    img_url = parse_image_urls(recipe, "recipetineats")
    if not img_url:  # If the first site fails, try the second
        img_url = parse_image_urls(recipe, "allrecipes")
    return img_url

df.loc[:19, "image_link"] = df.loc[:19, "title"].apply(get_image_link)

# Save the updated DataFrame
# df.to_csv("recipes_with_images.csv", index=False)

print("Image links added successfully!")

driver.quit()
```
